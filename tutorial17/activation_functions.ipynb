```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0, x)
x = np.array([-3, -1, 0, 2, 4])
print("ReLU:", relu(x))
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
x = np.array([-3, -1, 0, 2, 4])
print("Sigmoid:", sigmoid(x))
```

```python
def tanh(x):
    #x = np.array([-3, -1, 0, 2, 4])
    return np.tanh(x)
print("Tanh:", tanh(x))
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum()
x = np.array([1.0, 2.0, 3.0])
print("Softmax:", softmax(x))
```

```python
def plot_activation(x, y, title):
    plt.plot(x, y)
    plt.title(title)
    plt.grid(True)
    plt.show()

x = np.linspace(-10, 10, 100)
plot_activation(x, relu(x), "ReLU Activation")
plot_activation(x, sigmoid(x), "Sigmoid Activation")
plot_activation(x, tanh(x), "Tanh Activation")
```

```python
x = np.array([2.0, 1.0, 0.1])
y = softmax(x)
plt.bar(range(len(x)), y)
plt.title("Softmax Activation (probability distribution)")
plt.xlabel("Class")
plt.ylabel("Probability")
plt.xticks(range(len(x)))
plt.grid(True, axis='y')
plt.show()
```
